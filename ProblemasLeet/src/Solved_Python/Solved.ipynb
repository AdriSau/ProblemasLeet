### 11. Container With Most Water

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        max_area = 0
        left = 0  
        right = len(height)-1
        while(left != right):
            max_area = max(max_area, ((right - left) * min(height[left],height[right])))
            if (height[right] > height[left]):
                left+=1
            else:
                right-=1
        return max_area
```

### 15. 3Sum

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        res = []
        if len(nums) < 3:
            return res
        nums = sorted(nums)
        low = 0
        high = 0
        for i in range(len(nums)-2):
            if ((i != 0) and (nums[i] == nums[i-1])):
                continue
            base = nums[i]
            low = i+1
            high = len(nums)-1
            while(low != high):
                if ((base + nums[low] + nums[high] == 0) and ([base,nums[low],nums[high]] not in res)):
                    res.append([base,nums[low],nums[high]])
                    low+=1
                elif((base + nums[low] + nums[high] > 0)):
                    high-=1
                else:
                    low+=1
        return res
            
                    

            
            
```

### 22. Generate Parentheses

```python
class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        
        if (n == 1):
            return["()"]
        else: 
            return self.generar_apertura("(",n-1,n) + self.generar_cierre("(",n-1,n)

    def generar_apertura(self,res, aperturas, cierres):
        res += "("
        aperturas-=1
        if((aperturas == 0) &(cierres == 0)):
            return [res]
        else:
            exit = []
            if(cierres > aperturas):
                exit = exit +  self.generar_cierre(res,aperturas,cierres)
            if(aperturas > 0):
                exit = exit +  self.generar_apertura(res,aperturas,cierres)
            return exit

    def generar_cierre(self,res, aperturas, cierres):
        res += ")"
        cierres-=1
        if((aperturas == 0) &(cierres == 0)):
            return [res]
        else:
            exit = []
            if(cierres > aperturas):
                exit = exit +  self.generar_cierre(res,aperturas,cierres)
            if(aperturas > 0):
                exit = exit +  self.generar_apertura(res,aperturas,cierres)
            return exit
```

### 84. Largest Rectangle in Histogram

```python
class Solution(object):
    def largestRectangleArea(self, heights):
        """
        :type heights: List[int]
        :rtype: int
        """
        max_val = 0
        stack = []
        for num in range(len(heights)):
            if (len(stack) == 0):
                stack.append([0,heights[num]])
            else:
                if (heights[num] < stack[-1][1]):
                    while ((len(stack) > 0) and (heights[num]  < stack[-1][1])):
                        pop_it = stack.pop()
                        max_val = max(((num - pop_it[0]) * pop_it[1]), max_val)
                    if (len(stack) == 0): 
                        stack.append([0,heights[num]])
                    else:
                        stack.append([(pop_it[0]),heights[num]])
                else:
                    stack.append([num,heights[num]])
        while(len(stack) > 0):
            pop_it = stack.pop()
            max_val = max(((len(heights) - pop_it[0]) * pop_it[1]), max_val)
        return max_val
```

### 125. Valid Palindrome

```python
import re
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        s = s.lower()
        s = re.sub(r'[^a-zA-Z0-9]','',s)
        for i in range(len(s) // 2):
            if(s[i] != s[len(s)-1-i]):
                return False
        return True
```

### 128. Longest Consecutive Sequence

```python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if ((not nums) |( len(nums) == 0 )):
            return 0
        
        streak = 0
        current = 0
        prev = 0
        nums = sorted(nums)
        for i in range(len(nums)):
            if ((i == 0) | (nums[i] == prev+1 )):
                current+=1
            elif((nums[i] == prev )):
                prev = prev
            else:
                streak = max(current, streak)
                current = 1
            prev = nums[i]
        
        return max(current, streak)
            
```

```python
class Solution(object):
    def dailyTemperatures(self, temperatures):
        """
        :type temperatures: List[int]
        :rtype: List[int]
        """
        lis_sav = []
        res = [0] * len(temperatures)

        for i in range(len(temperatures)):
            if ((i == 0) or (len(lis_sav) == 0) or (temperatures[i] <= lis_sav[(len(lis_sav)-1)][0])):
                lis_sav.append([temperatures[i],i])
            else:
                while((lis_sav) and (temperatures[i] > lis_sav[(len(lis_sav)-1)][0])):
                    item = lis_sav.pop()
                    res[item[1]] = (i - item[1])
                lis_sav.append([temperatures[i],i])
                
        
        return res

            

        
```
